## Import Package

```python
# Import package as pandas does not ship along as native package
# use pip install pandas if not installed yet, or follow setup guide

import pandas as pd
import matplotlib.pyplot as plt
```

## A Simple Example: How many Pokemon are there for each type?

```python
dataframe = pd.read_csv("https://raw.githubusercontent.com/KianYang-Lee/pandas-tutorial/main/datasets/Pokemon.csv")
dataframe.head()
```

## Useful Attributes and Methods

```python
# Checking for overview of DataFrame
dataframe.info()
```

```python
# Checking for structure of DataFrame
# number of rows and columns
dataframe.shape
```

```python
# number of elements in total
dataframe.size
```

```python
# number of dimension
dataframe.ndim
```

```python
# Checking for data types
dataframe.dtypes
```

```python
dataframe.HP.dtype
```

```python
dataframe["HP"].dtype
```

```python
dataframe.HP.astype("float64")
```

```python
# Checking first and last n rows
dataframe.head(10)
```

```python
dataframe.tail()
```

```python
# Checking for indices
dataframe.index
```

```python
dataframe.columns
```

```python
# Validating uniqueness
dataframe.duplicated()
```

```python
dataframe.nunique()
```

```python
dataframe.Name.unique()
```

```python
# Fast counting
dataframe["Type 1"].value_counts()
```

```python
# Checking for missing values
dataframe.isnull()
```

```python
dataframe.isnull().sum()
```

```python
# Compute descriptive statistics
dataframe.describe()
```

```python
dataframe.Name.describe()
```

```python
# Compute correlation matrix
dataframe.corr()
```

```python
# Getting n-largest values for a column
dataframe.nlargest(10, "Attack")
```

```python
# Getting specific summative value on a Series
dataframe.Attack.mean()
```

```python
dataframe.HP.median()
```

```python
dataframe.Defense.max()
```
