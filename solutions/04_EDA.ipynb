# 04 - Exploratory Data Analysis (EDA)

## Import Package

```python
# Import package as pandas does not ship along as native package
# use pip install pandas if not installed yet, or follow setup guide

import pandas as pd
```

```python
dataframe = pd.read_csv("https://raw.githubusercontent.com/KianYang-Lee/pandas-tutorial/main/datasets/Pokemon.csv")
dataframe.head()
```

## Practice Section

Using the `DataFrame` containing only the First Generation of Pokemon, perform EDA to gather information on the following subjects:
- Details of `DataFrame`
- Index
- List of columns
- Descriptive statistics
- Counts for each category for all the categorical columns
- Correlation matrix 
- Top 5 Pokemon with highest speed

```python
first_gen_pokemon = dataframe[dataframe.Generation == 1]
```

```python
# - Details of `DataFrame`
first_gen_pokemon.info()
# - Index
first_gen_pokemon.index
# - List of columns
first_gen_pokemon.columns
# - Descriptive statistics
first_gen_pokemon.describe()
# - Counts for each category for all the categorical columns
# Get all 'object' data type columns
categorical_cols = first_gen_pokemon.select_dtypes(include=['object']).columns.to_list()
for col in categorical_cols:
    if first_gen_pokemon[col].value_counts().keys().__len__() < 20:
        print(first_gen_pokemon[col].value_counts())
# - Correlation matrix 
first_gen_pokemon.corr()
# - Top 5 Pokemon with highest speed
first_gen_pokemon.nlargest(5, "Speed")
```
